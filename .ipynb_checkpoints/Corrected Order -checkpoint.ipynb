A comparison for the statistics was established by creating a dataframe to compare the means of Grade I and Grade II Meningioma. The fold changes between the grades is also calculated and presented as a figure. 

```python
# create a dataframe

feat_means=pd.DataFrame({'grade_1':grade_1[:-1]['mean'], 'grade_2':grade_2[:-1]['mean']})

# calculate fold changes and add to dataframe

fc=(feat_means.grade_2 - feat_means.grade_1)/feat_means.grade_1
feat_means['fc']=fc

# calculate Log2 of fold change and add to data frame for better visualisation. 

log2_fc=np.log2(np.abs(fc))
feat_means['log2_fc']=log2_fc
feat_means=feat_means.reset_index().rename(columns={'Grade':'feature'})

# sort dataframe according to fold change

feat_means['abs']=abs(feat_means['fc'])

# Sort by fold absolute change

feat_means=feat_means.sort_values(by=['abs'],ascending=False).drop(columns=['abs'])
feat_means.head(10)
```

## Data Scaling
To ensure all features follow the same range, data scaling is complete, this allows for dominance to be avoided. Data was scaled used Standard Scaler.


```python
#Load in library for Standard Scaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

#Extract and remove Grade from df2

column_grade = df2["Grade"]
df2_no_grade = df2.drop(columns=["Grade"])

#Scale df2 without the Grade feature

scaled_df2 = sc.fit_transform(df2_no_grade)

#New df for scaled data with no grade

scaled_df2_no_grade = pd.DataFrame(scaled_df2, columns=df2_no_grade.columns)

#check

scaled_df2_no_grade
```

## Outlier Detection

Isolation forest was used on the scaled data to detect any outlier, a random seed of 8 was used and outliers were then filtered and removed from the scaled data. 


```python
from sklearn.ensemble import IsolationForest

#Isolation first

isolation_forest = IsolationForest(contamination=0.05, random_state=8)

#fit isolation model to the scaled data not containing grade feature

isolation_forest.fit(scaled_df2_no_grade)

#Predict, filter and remove outliers in the scaled data

predict_outliers = isolation_forest.predict(scaled_df2_no_grade)

filter_outliers = np.where(predict_outliers == -1)[0]

df2_no_grade_cleanscale = scaled_df2_no_grade.drop(scaled_df2_no_grade.index[filter_outliers])

#Ensure rows are also removed from the Grade feature

clean_grade_column = column_grade.drop(column_grade.index[filter_outliers])

#Concatenate (join) Grade feature to the clean scaled data

df2_yes_grade_cleanscale = pd.concat([df2_no_grade_cleanscale, clean_grade_column], axis=1)

#check

df2_yes_grade_cleanscale
```

```python
#Put grade as the first column for easier visualization

df2_cleaned = df2_yes_grade_cleanscale[["Grade"] + [col for col in df2_yes_grade_cleanscale.columns if col != "Grade"]]

#Check

df2_cleaned
```

Isolation forest was able to detect and remove 5 outliers from the data. 


## Data Splitting

Data was split with 70% training and 30% testing. 3 machine learning classiers (random forest, logistic regression and support vector machine) were used on the training dataset.

```python
#Load in library for data splitting

from sklearn.model_selection import train_test_split

#Then, split the dataset into the data and labels
#Alternative code:
#X = df2_cleaned.drop(columns=["Grade"])
#y = df2_cleaned["Grade"]

X,y = df2_cleaned.drop(columns = ['Grade']), df2_cleaned['Grade']

#Split the data into train and test sets with 70% training and 30% testing, a random seed was set for consistency purposes.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8)
```

```python
#check

print(X.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#Training dataset to train 3 ML classifiers:
#Train with Random Forest
#load in library for Random Forest

from sklearn.ensemble import RandomForestClassifier

#Set a random seed and train the classifer

rfcl = RandomForestClassifier(n_estimators=100, random_state=8)
rfcl.fit(X_train, y_train)

#Check accuracy

print("Accuracy on test data using random forest: {:.2f}".format(rfcl.score(X_test, y_test)))
```

```python
#Check Random forest classifer feature importance

from sklearn.inspection import permutation_importance

result = permutation_importance(rfcl, X_train, y_train, n_repeats=10, random_state=42)
perm_sorted_idx = result.importances_mean.argsort()

tree_importance_sorted_idx = np.argsort(rfcl.feature_importances_)
tree_indices = np.arange(0, len(rfcl.feature_importances_)) + 0.5

fig, (ax1) = plt.subplots(1, 1, figsize=(12, 8))
ax1.barh(tree_indices, rfcl.feature_importances_[tree_importance_sorted_idx], height=0.7)
ax1.set_yticks(tree_indices)
ax1.set_yticklabels(X_train.columns[tree_importance_sorted_idx])
ax1.set_ylim((0, len(rfcl.feature_importances_)))

fig.tight_layout()
plt.show()
```

```python
#Train data set to train logisitic regression classifier

#load in required libary

from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()

# Train data

logistic.fit(X_train, y_train)

#check accuracy

print("Accuracy on test data using logistic regression: {:.2f}".format(logistic.score(X_test, y_test)))
```

```python
#Train dataset to train support vector machine (SVM) classifier

#load in required library

from sklearn.svm import SVC

svm = SVC()

#Train data

svm.fit(X_train, y_train)

#check accuracy

print("Accuracy on test data using logistic regression: {:.2f}".format(svm.score(X_test, y_test)))
```

Random Forest presents with the highest accuracy of 0.85 whilst Logistic Regression has a low accuracy of 0.63.

```python
#Load in libraries

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
import matplotlib.pyplot as plt

# Define estimator for feature importance, method for normalised importance.

estimator = RandomForestClassifier(n_estimators = 100, max_depth = 7, min_samples_leaf = 2, min_samples_split = 2, random_state=220)

estimator.fit(X_train, y_train)

# get feature importances

embedded_feature_ranking = pd.DataFrame({'feature':X_train.columns,'importance':estimator.feature_importances_})

# create a dataFrame of feature importance

imp=pd.DataFrame({'feature':X_train.columns,'importance':estimator.feature_importances_}).sort_values(by='importance',ascending=False)


# get significant features

significant_features = imp[imp.importance >= 0.0]

#Plot features by rank

imp.plot.bar(x='feature', y='importance', rot=0,figsize=(22,8), color='blue')
plt.xticks(rotation='vertical',fontsize=11)
plt.yticks(fontsize=14)
plt.xlabel('')
plt.ylabel('Normalised importance',fontsize=14)
plt.legend(fontsize=14)
plt.title('Features Ranked by Normalised Importance',fontsize=16)
plt.tight_layout()
```

```python
#Correlation-based method for filter-based method on train data for feature ranking

#import libaries needed

import matplotlib.pyplot as plt
import seaborn as sns


# Calculate Pearson correlation coefficients between features and grade

correlation_coefficients = X_train.corrwith(y_train)

# Sort order of the absolute correlation coefficients

sorted_correlation_coefficients = correlation_coefficients.abs().sort_values(ascending=False)

# Select the top features

top_features = sorted_correlation_coefficients.index[:10] 

# Filter the training data to include only the selected features

X_train_filtered = X_train[top_features]

# Check the selected features

print("Selected features based on Pearson correlation coefficient:")
X_train_filtered.head()

```

```python
#Feature Ranking

from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import random

# Define the classifier for RFE

estimator = RandomForestClassifier(random_state = 8)

rfe = RFE(estimator, n_features_to_select=1)

# Fit RFE to the training data

rfe.fit(X_train, y_train)

# Rank features

rf_ranking = rfe.ranking_

# Feature names

features = X_train.columns

# Create a df for feature ranking

rfe_feature_ranking = pd.DataFrame({'feature': features, 'rank': rf_ranking})

# Plot the feature ranking from 1 to 48

ranked_ref = rfe_feature_ranking.sort_values(by='rank')
ranked_ref.plot.bar(x='feature', y='rank', rot=0, figsize=(22, 8), color='blue')
plt.xticks(rotation='vertical', fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('')
plt.ylabel('Rank', fontsize=14)
plt.legend(fontsize=14)
plt.title('Features Ranked by Importance (RFE)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#Elastic Net 

#Import library required

from sklearn.linear_model import ElasticNetCV

#create model with cv, cv= 5 is commonly used as it balances effiency and perforamnce estimate reliability.

elastic_net = ElasticNetCV(cv=5, random_state=8)

#Fit to training data

elastic_net.fit(X_train, y_train)

#get selected features

select_features = X_train.columns[elastic_net.coef_ != 0]

#print chosen feature names

print("Selected Features:")
print(select_features)
```

```python
#Check x_test, 19 features selected

X_test_select_features
```

```python
#RFECV
#Load in required libaries

from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFECV

rfe_cv = RFECV(estimator = LogisticRegression(), cv =5)

#fit RFECV model

rfe_cv.fit(X_train_select_features, y_train)

#Indices

select_feature_ind_rfecv = rfe_cv.support_

#Selected feature names

select_feature_rfecv = select_features[select_feature_ind_rfecv]

#display feature names

print("selected features after RFECV:")
print(select_feature_rfecv)
```

```python
#Display final number of selected features

print("Optimal number of features:", rfe_cv.n_features_)
```

```python
#filter X_train and X_test

X_train_selected = X_train_select_features[select_feature_rfecv]
X_test_selected = X_test_select_features[select_feature_rfecv]

X_train_selected
```

Using Elastic Net followed by RFECV with Logistic regression as its base, feature selection was achieved and 16 features were obtained. 

## Train Model

Accuracy of performance was calcuated for the 3 classifiers on the training set. Information gathered was then plotted as boxplots prior to model evaluation.

```python
#Import required libaries to train models

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Logistic Regression 

lr = LogisticRegression()
lr_scores = cross_val_score(lr, X_train_selected, y_train, cv=5, scoring='accuracy')
print("Logistic Regression Mean Accuracy:", lr_scores.mean())

# Random Fores

rf = RandomForestClassifier()
rf_scores = cross_val_score(rf, X_train_selected, y_train, cv=5, scoring='accuracy')
print("Random Forest Mean Accuracy:", rf_scores.mean())

# Support Vector Machine (SVM)

svm = SVC()
svm_scores = cross_val_score(svm, X_train_selected, y_train, cv=5, scoring='accuracy')
print("SVM Mean Accuracy:", svm_scores.mean())
```

```python
#Load in libaries required

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Define the 3 models

models = [
    ('LR', LogisticRegression()),
    ('RF', RandomForestClassifier()),
    ('SVM', SVC())
]

# Perform cross-validation and collect results for accuracy scoring

results = []
names = []
for name, model in models:
    scores = cross_val_score(model, X_train_selected, y_train, cv=5, scoring='accuracy')
    results.append(scores)
    names.append(name)

# Create DF for the results

df = pd.DataFrame(results).T
df.columns = names

# Display mean accuracy scores

print("Accuracy Scores:")
print(df.mean())

# Plot boxplots

plt.figure(figsize=(10, 6))
plt.boxplot(df.values, labels=df.columns)
plt.title('Machine Learning Algorithm Comparison for LR, RF and SVM Algorithms Against Accuracy')
plt.xlabel('Machine Learning Algorithm')
plt.ylabel('Accuracy')
plt.show()
```

## Model Evaluation

Models were evaluated using performance scores: accuracy, precision, recall, f1 and AUC prior to hyperparameter tuning for comparison purposes.

```python
#Load in required libary for model evaluation. Performance scores were used for evaluation

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Train the models (SVM, LR and RF) on the training dataset using th 16 selected features
lr.fit(X_train_selected, y_train)
rf.fit(X_train_selected, y_train)
svm.fit(X_train_selected, y_train)

# Evaluate the models using the test set

lr_pred = lr.predict(X_test_selected)
rf_pred = rf.predict(X_test_selected)
svm_pred = svm.predict(X_test_selected)

# Calculate performance metrics

lr_accuracy = accuracy_score(y_test, lr_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
svm_accuracy = accuracy_score(y_test, svm_pred)

lr_precision = precision_score(y_test, lr_pred)
rf_precision = precision_score(y_test, rf_pred)
svm_precision = precision_score(y_test, svm_pred)

lr_recall = recall_score(y_test, lr_pred)
rf_recall = recall_score(y_test, rf_pred)
svm_recall = recall_score(y_test, svm_pred)

lr_f1 = f1_score(y_test, lr_pred)
rf_f1 = f1_score(y_test, rf_pred)
svm_f1 = f1_score(y_test, svm_pred)

lr_auc = roc_auc_score(y_test, lr_pred)
rf_auc = roc_auc_score(y_test, rf_pred)
svm_auc = roc_auc_score(y_test, svm_pred)

# Display the performance measures

print("Logistic Regression:")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1-score:", lr_f1)
print("AUC:", lr_auc)
print()

print("Random Forest:")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1-score:", rf_f1)
print("AUC:", rf_auc)
print()

print("Support Vector Machine:")
print("Accuracy:", svm_accuracy)
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1-score:", svm_f1)
print("AUC:", svm_auc)
```

## Hyperparameter Tuning
Here, hyperparameter tuning using grid-search was conducted. The models were then updated and re-evaluated. A ROC curve was made for the best perfoming classifier.

```python
#Load in libaries required for grid search of all 3 classifiers.

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Define the grids for the models

lr_grid = {'C': [0.1, 1, 10, 100], 'penalty': ['l2']}
rf_grid = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
svm_grid = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']}

# Conduct grid search with cross-validation (cv =5) for each model

lr_gridsearch = GridSearchCV(LogisticRegression(), lr_grid, cv=5, scoring='accuracy')
lr_gridsearch.fit(X_train_selected, y_train)

rf_gridsearch = GridSearchCV(RandomForestClassifier(), rf_grid, cv=5, scoring='accuracy')
rf_gridsearch.fit(X_train_selected, y_train)

svm_gridsearch = GridSearchCV(SVC(), svm_grid, cv=5, scoring='accuracy')
svm_gridsearch.fit(X_train_selected, y_train)

# Update models with the best hyperparameters

lr_hyper_model = lr_gridsearch.best_estimator_
rf_hyper_model = rf_gridsearch.best_estimator_
svm_hyper_model = svm_gridsearch.best_estimator_

# Re-evaluate the tuned models on the test set

lr_test_predict = lr_hyper_model.predict(X_test_selected)
rf_test_predict = rf_hyper_model.predict(X_test_selected)
svm_test_predict = svm_hyper_model.predict(X_test_selected)


# Re-calculate the performance metrics

lr_test_acc = accuracy_score(y_test, lr_test_predict)
rf_test_acc = accuracy_score(y_test, rf_test_predict)
svm_test_acc = accuracy_score(y_test, svm_test_predict)

lr_test_precsn = precision_score(y_test, lr_test_predict)
rf_test_precsn = precision_score(y_test, rf_test_predict)
svm_test_precsn = precision_score(y_test, svm_test_predict)

lr_test_recall = recall_score(y_test, lr_test_predict)
rf_test_recall = recall_score(y_test, rf_test_predict)
svm_test_recall = recall_score(y_test, svm_test_predict)

lr_f1_test = f1_score(y_test, lr_test_predict)
rf_f1_test = f1_score(y_test, rf_test_predict)
svm_f1_test = f1_score(y_test, svm_test_predict)

lr_auc_test = roc_auc_score(y_test, lr_test_predict)
rf_auc_test = roc_auc_score(y_test, rf_test_predict)
svm_auc_test = roc_auc_score(y_test, svm_test_predict)

# Display the tuned performance metrics

print("Hyperparameter tuned: Logistic Regression:")
print("Accuracy:", lr_test_acc)
print("Precision:", lr_test_precsn)
print("Recall:", lr_test_recall)
print("F1-score:", lr_f1_test)
print("AUC:", lr_auc_test)
print()

print("Hyperparameter tuned: Random Forest:")
print("Accuracy:", rf_test_acc)
print("Precision:", rf_test_precsn)
print("Recall:", rf_test_recall)
print("F1-score:", rf_f1_test)
print("AUC:", rf_auc_test)
print()

print("Hyperparameter tuned: Support Vector Machine:")
print("Accuracy:", svm_test_acc)
print("Precision:", svm_test_precsn)
print("Recall:", svm_test_recall)
print("F1-score:", svm_f1_test)
print("AUC:", svm_auc_test)
```

```python
#ROC Graph Curve for the best performing model

from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predict probabilities for best performing model, in this case, Logistic Regression

lr_probs = lr_hyper_model.predict_proba(X_test_selected)[:, 1]
#rf_probs = rf_hyper_model.predict_proba(X_test_selected)[:, 1]
#svm_probs = svm_hyper_model.predict_proba(X_test_selected)[:, 1]

# Calculate the false positive and true positive rates

lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
#rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
#svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_probs)

# Calculate the ROC AUC score 

lr_auc = roc_auc_score(y_test, lr_probs)
#rf_auc = roc_auc_score(y_test, rf_probs)
#svm_auc = roc_auc_score(y_test, svm_probs)

# Plot ROC curve 

plt.figure(figsize=(10, 10))
plt.plot(lr_fpr, lr_tpr, label='Logistic Regression (AUC = %0.2f)' % lr_auc)
#plt.plot(rf_fpr, rf_tpr, label='Random Forest (AUC = %0.2f)' % rf_auc)
#plt.plot(svm_fpr, svm_tpr, label='SVM (AUC = %0.2f)' % svm_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression Model')
plt.legend(loc="lower right")
plt.show()
```
