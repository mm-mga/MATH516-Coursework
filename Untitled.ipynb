```python
#Load in the dataset 'Dataset_01.csv'
#df = pd.read_csv('C:\\Users\\mmmiguezgarciaafonso\\OneDrive - University of Plymouth\\MSc 2023-2024\Machine Learning and Artificial intelligence for Healthcare (MATH516)\\coursework\\Dataset _01.csv')
df = pd.read_csv('C:\\Users\\mmmig\OneDrive - University of Plymouth\\MSc 2023-2024\\Machine Learning and Artificial intelligence for Healthcare (MATH516)\\coursework\\Dataset _01.csv')
#Print to check
print(df)
```

```python
#Removal of non-numeric features

#check for numeric variables
def numeric_wanted(column):
    return pd.to_numeric(column, errors = 'coerce').notnull().all()

wanted_variables = [col for col in df.columns if numeric_wanted(df[col])]

#dataframe for numeric variables only
df2 = df[wanted_variables]

print(df2)
```

```python
#Descriptive statistics for Grade 1 Meningioma
#Remove the index, Grade, from the output
grade_1 = df2[df2['Grade'] == 0].drop(columns = ['Grade']).describe().T

#Show first 5 features
grade_1.head(5)

#Descriptive Statistics for Grade 2 Meningioma
#Remove the index, Grade from the output
grade_2 = df2[df2['Grade'] == 1].drop(columns = ['Grade']).describe().T

#Show first 5 features
grade_2.head(5)
```

```python
# create dataframe
feat_means=pd.DataFrame({'grade_1':grade_1[:-1]['mean'], 'grade_2':grade_2[:-1]['mean']})

# calculate fold changes and add to dataframe
fc=(feat_means.grade_2 - feat_means.grade_1)/feat_means.grade_1
feat_means['fc']=fc

# calculate Log2 of fold change and add to data frame
log2_fc=np.log2(np.abs(fc))
feat_means['log2_fc']=log2_fc
feat_means=feat_means.reset_index().rename(columns={'Grade':'feature'})

# sort dataframe according to fold change
feat_means['abs']=abs(feat_means['fc'])

# Sort by fold absolute change
feat_means=feat_means.sort_values(by=['abs'],ascending=False).drop(columns=['abs'])
feat_means.head(10)
```

```python
import matplotlib.pyplot as plt

# Create the dataframe
feat_means = pd.DataFrame({'grade_1': grade_1[:-1]['mean'], 'grade_2': grade_2[:-1]['mean']})

# Calculate fold changes and add to the dataframe
fc = (feat_means['grade_2'] - feat_means['grade_1']) / feat_means['grade_1']
feat_means['fc'] = fc

# Calculate Log2 of fold change and add to dataframe
log2_fc = np.log2(np.abs(fc))
feat_means['log2_fc'] = log2_fc
feat_means = feat_means.reset_index().rename(columns={'index': 'feature'})

# Sort dataframe according to fold change
feat_means['abs'] = abs(feat_means['fc'])
feat_means = feat_means.sort_values(by='abs', ascending=False).drop(columns='abs').head(10)

# Plot
plt.figure(figsize=(10, 10))
plt.bar(feat_means['feature'], feat_means['log2_fc'], color='goldenrod', width=0.5)
plt.xticks(rotation='vertical')
plt.xlabel('Feature')
plt.ylabel('Fold Change (Log2)')
plt.tight_layout()

# Save plot
plt.savefig('fold2change.jpeg')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the "Grade" column
grade_column = df2["Grade"]

# Drop the "Grade" column from the DataFrame
df_without_grade = df2.drop(columns=["Grade"])

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data without the "Grade" column
scaled_data = scaler.fit_transform(df_without_grade)

# Convert the scaled data back to a DataFrame
scaled_df_without_grade = pd.DataFrame(scaled_data, columns=df_without_grade.columns)

# Concatenate the "Grade" column back to the scaled DataFrame
scaled_df_with_grade = pd.concat([scaled_df_without_grade, grade_column], axis=1)

# Display the scaled DataFrame with the "Grade" column
scaled_df_with_grade
```

```python
from sklearn.ensemble import IsolationForest

# Drop the "Grade" column from the DataFrame before applying Isolation Forest
df_without_grade = df2.drop(columns=["Grade"])

# Initialize the Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05, random_state=8)

# Fit the model to the data without the "Grade" column
isolation_forest.fit(df_without_grade)

# Predict outliers
outlier_prediction = isolation_forest.predict(df_without_grade)

# Filter outliers
outlier_indices = np.where(outlier_prediction == -1)[0]

# Remove outlier points from the DataFrame
cleaned_df_without_grade = df_without_grade.drop(df_without_grade.index[outlier_indices])

# Also remove corresponding rows from the "Grade" column
cleaned_grade_column = df2["Grade"].drop(df2.index[outlier_indices])

# Concatenate the "Grade" column back to the DataFrame
cleaned_df_with_grade = pd.concat([cleaned_df_without_grade, cleaned_grade_column], axis=1)

# Display the cleaned DataFrame
cleaned_df_with_grade
```
